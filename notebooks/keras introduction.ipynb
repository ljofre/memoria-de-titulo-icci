La metodología que usaremos para hacer nuestra primera red neuronal con Keras es

* Lectura de datos
* Definir el modelo
* Compilar el modelo
* Ajustar el modelo
* Evaluar el modelo
* Ensamblar todo junto


```python
#keras_first_network.py

from keras.models import Sequential
from keras.layers import Dense
import numpy
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

```python
# load pima indians dataset
dataset = numpy.loadtxt("./keras/pima-indians-diabetes.data", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
print "número de features: ", X.shape[0]
print "número de casos: ", X.shape[1]
```

Crearemos un modelo secuencial y le iremos agregando capas hasta que nos sintamos felices.

Se ve que hay 8 Features y 786 casos, ahora crearemos nuestra red, o sea, 
definiremos el modelo para predecir el valor de $y_i$ a partir de $x_i$

La primera cosa con la que nos debemos asegurar es que tegamos la cantidad correcta de entradas, que debe ser igual
a la cantidad de Features. Usaremos red con tres capas ocultas. Una red de con capas completamente conectadas es implementad usando la clase **Dense**. Especificaremos en número de neuronas en la capa como primer argumento, el método de inicialización como segundo argumento como init y especificar la función de activación usando el argumento artivación.

En este caso, inicializaremos los pesos con un número aleatorio pequeño generado $$\theta \sim \text{Uniform}\left(0,0.05\right)$$ que es la opción por defecto de Keras.

La primera capa tiene 12 neuronas y espara los 8 features de entrada, la segunda capa tiene 8 neuronas y la tercera capa tiene una neurona, que es la que predice si tiene diabetes o no.

```python
model = Sequential()
model.add(Dense(12, input_dim=8, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
```

Compilar el modelo
==
Ahora que ya la red está definida podemos compilarla. La compilación usa las librerías numéricas dadas por theano o TensorFlow. Theano o tensorFlow selecciona automáticamente la mejor manera de representar la red para entrenarla.

Para entrenarla necesitamos definir adicionalmente algunas propiedades para cuando se entrene la red. Recuerde que entrenar la red significa encontrar los mejores pesos para hacer las predicciones para los datos. 

En este caso usaremos "logaritmic loss", la cual es bastante util para optimización de variables binarias. En Keras está definida por "binary_crossentropy". También usaremos un algoritmo de desenco de gradiente llamado "adam" por la razón de que es el método por defecto.
http://arxiv.org/pdf/1412.6980v8.pdf



```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

Ajustar el modelo
==

Hemos definido el modelo y lo hemos compilado y está listo para un cálculo eficiente de los pesos en función de los datos disponibles (entrenamiento).

Podemos entrenar nuestro modelo usando la función fit()

El proceso de entrenamiento implica un conjunto de iteraciones a través del set de datos llamado "épocas". podemos definir el número de épocas usando el argumento nb_epoch.

También podemos setear el número de instancias que serán evaluadas antes de la actualización de los pesos usando la variable batch_size.

Para este problema, usaremos un pequeño numero de iteraciones (150) y un numero relativamente pequeño de batch size = 10. Uno debe serleccionarlos experimentalmente mediante ensayo y error.

```python
# Fit the model
model.fit(X, Y, nb_epoch=150, batch_size=10)
```

Evaluar el modelo
==

```python
# evaluate the model
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
